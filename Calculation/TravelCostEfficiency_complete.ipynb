### 计算travel cost = travelTime * valueTime

```python
len(data['networks'])
```

```python
#存储networks中最后一组数据，提取endVehicles，将在最后一帧时间戳中，已经结束的车辆，存入EndCars中。
EndCars = data['networks'][-1]['endVehicles']
```

```python
# 计算travel cost 并存入 dictionary中 （例）
i = 0
TravelCost = {}
for key, value in EndCars.items():
    print(key, '->', value)
    print("value time: ", value['valueTime'], "travel time:", value['finishTs']-value['startTs'])
    print("Travel Cost: ", value['valueTime'] * (value['finishTs']-value['startTs']))
    TravelCost[value['id']] = value['valueTime'] * (value['finishTs']-value['startTs'])
    i = i + 1
    if i > 1 :
        break
print(TravelCost)
```

```python
# 计算travel cost 并存入 dictionary中 （全部存入）
TravelCost = {}
TravelTime = {}
for key, value in EndCars.items():
    TravelCost[value['id']] = value['valueTime'] * (value['finishTs']-value['startTs'])
    TravelTime[value['id']] = (value['finishTs']-value['startTs'])
print(len(TravelCost))
print(TravelCost)
print(TravelTime)
```

```python
# 将dictionary 转换为 dataframe，利用pandas
print(type(TravelCost), type(TravelTime))
TravelCost_frame = pd.DataFrame.from_dict(TravelCost, orient='index', columns = ['TravelCost/s'])
TravelTime_frame = pd.DataFrame.from_dict(TravelTime, orient='index', columns = ['TravelTime/s'])
```

```python
TravelCost_frame['TravelCost/h'] = TravelCost_frame['TravelCost/s'] / 3600
TravelTime_frame['TravelTime/h'] = TravelTime_frame['TravelTime/s'] / 3600
print(TravelCost_frame.head())
print("number of rows: ", len(TravelCost_frame))
print(TravelTime_frame.head())
print("number of rows: ", len(TravelCost_frame))
```

```python
#将上面两组dataframe结合在一起
TravelEfficiency = pd.concat([TravelCost_frame, TravelTime_frame], axis=1, sort=False)
print(TravelEfficiency.head())
```

```python
def getVOT(data):
    EndCars = data['networks'][-1]['endVehicles']
    ValueTime = {}
    for key, value in EndCars.items():
        ValueTime[value['id']] = value['valueTime']
    ValueTime_frame = pd.DataFrame.from_dict(ValueTime, orient='index', columns = ['ValueTime'])
    return ValueTime_frame
```

```python
def getEndVehicleInfo(data):
    EndCars = data['networks'][-1]['endVehicles']
    EndCars_frame = pd.DataFrame.from_dict(EndCars, orient='index')
    EndCars_frame = EndCars_frame.sort_values(by=['id'])
    return EndCars_frame    
```

```python
def TravelTimeCalculation(listNoCharge, listCharge, distributionName):
    NoCharge = listNoCharge[-1]
    Charge = listCharge[-1]
    print("No charge scenario => Read in document with path: ", NoCharge)
    print("Charge scenario => Read in document with path: ", Charge)
    dataNo = load_data(NoCharge)
    dataCh = load_data(Charge)
    out_noCharge = getEndVehicleInfo(dataNo)
    out_charge = getEndVehicleInfo(dataCh)
    print("vehicle information can be seen bellow: ")
    print("No Charge scenario: ")
    print(out_noCharge.head())
    print("Charge scenario: ")
    print(out_charge.head())
    out_all = pd.DataFrame()
    out_all['id'] = out_noCharge['id']
    out_all['departureTs'] = out_noCharge['startTs']
    out_all['departureTs_c'] = out_charge['startTs']
    out_all['VOT'] = out_noCharge['valueTime']
    out_all['VOT_c'] = out_charge['valueTime']
    out_all['type'] = out_noCharge['type']
    out_all['type_c'] = out_charge['type']
    #得到的值为正值，travel time花的更少
    out_all['TravelTimeDifference(h)'] = ((out_noCharge['finishTs'] - out_noCharge['startTs']) - (out_charge['finishTs'] - out_charge['startTs']))/3600
    outPath = 'D:/ChargingSystem/Calculation/output/' + distributionName + '_travelTime.csv' #get unique file name
    out_all.to_csv (outPath, index = None, header=True)
    print("Data has been generated to the path: D:/ChargingSystem/Calculation/output")    
```

```python
def calculateTT(data):
    EndCars = data['networks'][-1]['endVehicles']
    
    TravelCost = {}
    TravelTime = {}
    
    for key, value in EndCars.items():
        TravelCost[value['id']] = value['valueTime'] * (value['finishTs']-value['startTs'])
        TravelTime[value['id']] = (value['finishTs']-value['startTs'])
    
    TravelCost_frame = pd.DataFrame.from_dict(TravelCost, orient='index', columns = ['TravelCost/s'])
    TravelTime_frame = pd.DataFrame.from_dict(TravelTime, orient='index', columns = ['TravelTime/s'])
    TravelCost_frame['TravelCost/h'] = TravelCost_frame['TravelCost/s'] / 3600
    TravelTime_frame['TravelTime/h'] = TravelTime_frame['TravelTime/s'] / 3600
    TravelEfficiency = pd.concat([TravelCost_frame, TravelTime_frame], axis=1, sort=False)
    return TravelEfficiency
```

```python
def load_data(path):
    with open(path) as f:
        data = json.load(f)
    return data
```

```python
def travelEfficiency(listName):
    '''
    This function is to calculate the travel time and travel cost from the output datasets.
    Example:
    
    List = ["random_vol_sim1_False_0_output.json", "random_vol_sim1_False_1_output.json", "random_vol_sim1_False_2_output.json"]
    travelEfficiency(List)
    
    Output:
    
    Read in document with path:  D:/ChargingSystem/meso_v2.0/visualization/random_vol_sim1_False_2_output.json
    The calculated travel cost and travel time can be seen bellow: 
       TravelCost/s  TravelCost/h  TravelTime/s  TravelTime/h
    1  10214.931341      2.837481         192.0      0.053333
    2   7443.735768      2.067704         191.0      0.053056
    3   9939.445700      2.760957         190.0      0.052778
    4   7474.197550      2.076166         182.0      0.050556
    5   9627.004625      2.674168         182.0      0.050556
    
    In this function, we only read the very last document in the list, since the last document stores the finformation.
    '''
    file = listName[-1]
    print("Read in document with path: ", file)
    data1 = load_data(file)
    out_data1 = calculateTT(data1)
    print("The calculated travel cost and travel time can be seen bellow: ")
    print(out_data1.head())
    outPath = 'D:/ChargingSystem/Calculation/output/' + file.split('/')[-1].split('.')[0] + '_TravelEfficiency.csv' #get unique file name
    out_data1.to_csv (outPath, index = None, header=True)
    print("Data has been generated to the path: D:/ChargingSystem/Calculation/output")
```
